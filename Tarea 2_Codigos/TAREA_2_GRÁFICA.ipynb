```python
# Números divisibles por 5 y 100 + Bip
for x in range(1, 101):
    if x % 5 == 0:
        if x % 10 == 0:
            print(f"{x} Bip Bip")
        else:
            print(f"{x} Bip")
```

Se agrega un rango, se tiene sumo cuidado de considerar la posición 0 (por eso la posición final en el "for" es 101).

Se usa el % para considerar el resto de las divisiones, considerando que los números deben dar un resto 0 al ser divididos por 10, mostrándose un "Bip Bip". Para el resto de los casos (los números divisibles por 5), se agrega "Bip"

```python
# puntaje gamer
puntaje1 = int(input("Primer puntaje: "))
puntaje2 = int(input("Segundo puntaje: "))

suma = puntaje1 + puntaje2

if suma < 50:
    print("Resultado: Nivel principiante")
elif suma >= 50 and suma < 80:
    print("Resultado: Nivel intermedio")
else:
    print("Resultado: Nivel pro")
```

```python
# música y baile
edad = int(input("Ingresa tu edad: "))
le_gusta_musica = input("¿Te gusta la música? (si-no): ").lower() == "si"
le_gusta_bailar = input("¿Te gusta bailar? (si-no): ").lower() == "si"

if edad >= 18 and le_gusta_musica and le_gusta_bailar:
    print("Eres una estrella de la fiesta")
elif le_gusta_musica and not le_gusta_bailar:
    print("Te gusta la música, pero no tanto bailar")
else:
    print("Tienes tu propio estilo")
```
